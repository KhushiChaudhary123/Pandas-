```python
# Concatenate and Merge DataFrame(JOINS)
```

```python
df1 = pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4],'Score':[88,96,77,79]})
df2
```

```python
pd.concat([df1,df2],axis =0 ) # addition of rows vertically
```

```python
pd.concat([df1,df2],axis =1 )  # addition of rows horizontally
```

```python
#  merge = join
```

```python
pd.merge(df1,df2, how = 'inner' , on = 'ID')
```

```python
pd.merge(df1,df2, how = 'left' , on = 'ID')
```

```python
pd.merge(df1,df2, how = 'inner' , left_on = 'ID',right_on = 'ID')
```
